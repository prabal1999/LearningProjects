```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('7_1188.jpg', cv2.IMREAD_GRAYSCALE)
_,thres = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
cntrs,_ = cv2.findContours(thres, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
# final = cv2.drawContours(thres, cntrs, -1, (0,255,0), 10)
print(cntrs)

maxm = 0
for cnt in cntrs:
    x,y,w,h = cv2.boundingRect(cnt)
    maxm = max(w*h, maxm)
    if w*h == maxm:
        x_max = x
        y_max = y
        h_max = h
        w_max = w
cv2.rectangle(thres, (x_max,y_max), (x_max+w_max, y_max+h_max), (255,255,255), 1)

print(x,y,w,h)
print(x_max, y_max, w_max, h_max)
plt.imshow(thres, cmap='gray')
plt.show()
# cv2.waitKey(0)
# cnt = sorted(cntrs, key=lambda ctr: cv2.boundingRect(ctr)[0])
# cv2.imshow('window', thres)
# cv2.waitKey(0)
```

```python
im_crop = thres[y_max:y_max+h_max, x_max:x_max+w_max]
im_crop = cv2.resize(im_crop, (28,28))
feature_img = cv2.resize(im_crop, (784,1))
feature_img

# plt.imshow(feature_img, cmap='gray')
# plt.show()
```
